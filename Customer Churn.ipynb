#### <b>Bedeutung der Kundenabwanderungsvorhersage im Telekommunikationssektor</b>
<span style="font-size:16px;">Die Prognose der Kundenabwanderung - auch Churn Prediction genannt - ermöglicht es Unternehmen, gefährdete Kundinnen und Kunden frühzeitig zu identifizieren und gezielte Massnahmen zur Kundenrückgewinnung oder -bindung zu ergreifen.

<span style="font-size:16px;">Mit Hilfe moderner Analyseverfahren, insbesondere durch den Einsatz von künstlicher Intelligenz und maschinellem Lernen, können Muster im Kundenverhalten erkannt werden, die auf eine bevorstehende Abwanderung hindeuten.

<span style="font-size:16px;">Durch den gezielten Einsatz dieser Informationen können Telekommunikationsunternehmen nicht nur die Kundenzufriedenheit steigern, sondern auch langfristig ihre Umsätze stabilisieren und ihre Marktposition stärken
<a id = "3" ></a>
#### <b>Strategien zur Reduzierung der Kundenabwanderung</b>
<span style="font-size:16px;">Die Verringerung der Kundenabwanderung (Churn) ist für den langfristigen Erfolg eines Unternehmens entscheidend. Hier einige Strategien, um die Kundenabwanderung zu minimieren:
* <b>Verbesserung des Kundenservices</b>
  * Schnelle Reaktionszeiten und ein freundlicher Service erhöhen die Kundenzufriedenheit und beugen der Abwanderung vor.
* <b>Personalisierung</b>
  * Kunden erwarten Angebote, die auf sie zugeschnitten sind. Durch personalisierte Empfehlungen und Rabatte fühlen sich Kunden wertgeschätzt und bleiben eher treu.
* <b>Proaktive Kundenbindung</b>
  * Unternehmen sollten regelmäßig Feedback von Kunden einholen und proaktive Maßnahmen wie Treueprogramme oder exklusive Angebote anbieten.
* <b>Produkt- und Servicequalität</b>
  * Eine hohe Produkt- und Servicequalität ist der Schlüssel zur langfristigen Kundenbindung.
* <b>Frühzeitiges Erkennen von Abwanderungsrisiken</b>
  * Mit Datenanalysen und maschinellem Lernen können Unternehmen gefährdete Kunden frühzeitig erkennen und gezielte Maßnahmen ergreifen.-
<a id = "4" ></a>
#### <b> Ziel der Analyse</b>
<span style="font-size:16px;">Ziel dieses Projektes ist es, mit Hilfe explorativer Datenanalysen und prädiktiver Modellierung ein besseres Verständnis der Ursachen von Kundenabwanderung in der Telekommunikationsbranche zu erlangen. Dabei sollen unter anderem folgende Fragen untersucht werden:
* Wie hoch ist der prozentuale Anteil an abgewanderten Kund:innen im Vergleich zu jenen, die weiterhin aktiv Dienstleistungen nutzen?
* Gibt es geschlechtsspezifische Muster bei der Abwanderung?
* Welche Dienstleistungen sind aus Unternehmenssicht besonders profitabel?
* Welche Merkmale und Dienstleistungsangebote haben den höchsten wirtschaftlichen Nutzen?
* Darüber hinaus sollen im Rahmen der Analyse weitere relevante Fragestellungen identifiziert und untersucht werden, die sich aus den Daten ergeben.

#### <b> KNN Classifikation</b>
<a id = "knn" ></a>

```python
k = 3
#Train Model and Predict  
knn_classifier = KNeighborsClassifier(n_neighbors=k)
knn_model = knn_classifier.fit(X_train,y_train)
```

* wir können mit einem kleinen Wert für k beginnen, beispielsweise k = 3.

```python
yhat = knn_model.predict(X_test)
print("Test set Accuracy: ", accuracy_score(y_test, yhat))
```

* Test set Accuracy ist  0.738388625592417

```python
print(classification_report(y_test, yhat))
```

```python
k = 8
#Train Model and Predict  
knn_classifier = KNeighborsClassifier(n_neighbors=k)
knn_model = knn_classifier.fit(X_train,y_train)
yhat = knn_model.predict(X_test)
print("Test set Accuracy: ", accuracy_score(y_test, yhat))
```

* wir haben die k erhöht die Accuracy ist verbessert worden. 

```python
#Die Wahl des richtigen Wertes von k
Ks = 50
acc = np.zeros((Ks))
std_acc = np.zeros((Ks))
for n in range(1,Ks+1):
    #Train Model and Predict  
    knn_model_n = KNeighborsClassifier(n_neighbors = n).fit(X_train,y_train)
    yhat = knn_model_n.predict(X_test)
    acc[n-1] = accuracy_score(y_test, yhat)
    std_acc[n-1] = np.std(yhat==y_test)/np.sqrt(yhat.shape[0])
```

```python
plt.plot(range(1,Ks+1),acc,'g')
plt.fill_between(range(1,Ks+1),acc - 1 * std_acc,acc + 1 * std_acc, alpha=0.10)
plt.legend(('Accuracy value', 'Standard Deviation'))
plt.ylabel('Model Accuracy')
plt.xlabel('Number of Neighbors (K)')
plt.tight_layout()
plt.show()
```

* Der beste K-Wert ist 11, wie aus der Abbildung ersichtlich ist.

```python
k = 11
#Train Model and Predict  
knn_classifier = KNeighborsClassifier(n_neighbors=k)
knn_model = knn_classifier.fit(X_train,y_train)
yhat = knn_model.predict(X_test)
print("Test set Accuracy: ", accuracy_score(y_test, yhat))
```

```python
print(classification_report(y_test, yhat))
```

* Das Modell erkennt Kunden, die NICHT kündigen, schon ziemlich gut 

* Für Churn-Kunden (Klasse 1) ist die Performance noch eher schwach
